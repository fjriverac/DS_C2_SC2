```python
import pandas as pd
# a) Extraccion de la informacion del archivo csv en un Datafrane
# a.1) Conversion del campo de fechas en tipo datetime
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])

# Ignorar valores nulos
country_vaccinations.notna()

# b) Mostrar estructura del Dataframe y el Tipo de los Datos de cada columna
print('\n*** b) Estructura del Dataframe y Tipo de Datos de cada Columna ***\n')
country_vaccinations.info()

# c) Agrupar por Farmaceutica y contabilizar cantidad de vacunas de cada una de ellas
country_vaccinations['vaccines'].str.split()
country_vaccinations_companias=country_vaccinations['vaccines'].groupby(country_vaccinations['vaccines']).count()
print('\nc) *** Cantidad de vacunas por compa√±ias: ***\n',country_vaccinations_companias)

# d) Calculo del total de vacunas aplicadas a nivel mundial a partir de la columna Daily_Vaccinations
print('\nd) *** Total de vacunas aplicadas a nivel mundial: ',format(country_vaccinations['daily_vaccinations'].sum(), '0,.0f'))

# e) Calculo del promedio de vacunas aplicadas por pais
promedio_pais=country_vaccinations[['country','daily_vaccinations']].groupby('country').mean()
print('\ne) *** Promedio de vacunas aplicada por Pais ***',promedio_pais)
vacunas_290121=country_vaccinations[['date','daily_vaccinations']]
vacunas_290121=vacunas_290121.query('date=="1/29/2021"')

# f) Calculo del total de vacunas a nivel mundial del dia 29/01/21
print('\nf) *** Total de vacunas aplicadas a nivel mundial el dia 29/01/21: ',format(vacunas_290121['daily_vaccinations'].sum(),'0,.0f'))

# g) Crear Dataframe conDiferencias a partir del Dataframe Original
conDiferencias=country_vaccinations

# g.1) Insertar Columna Derivada con la diferencia entre campos daily_vaccinations y daily_vaccinations_raw
conDiferencias['diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
print('\ng) *** Columna Derivada con la diferencia entre campos daily_vaccinations y daily_vaccinations_raw\n', conDiferencias)

# h) Obtener el per√≠odo de tiempo entre el registro con fecha mas reciente y el registro con fecha mas antigua
print('\nh) *** Diferencia en dias entre el registro mas reciente y el mas antiguo: ',conDiferencias['date'].max()-conDiferencias['date'].min())

# i.1) Crear Dataframe conCantidad a partir del Dataframe Original
conCantidad=country_vaccinations

# i.2) Insertar Columna Derivada con la cantidad de vacunas utilizadas por Farmac√©utica
conCantidad['canVac']=conCantidad['vaccines'].str.split(', ').str.len()
print('\nh) *** Cantidad de vacunas utilizadas por Farmac√©utica:\n',conCantidad['canVac'])

# j) Crear Dataframe antes20 con los registros previos al 20/12/2020
antes20=country_vaccinations[country_vaccinations['date']<'12/20/2020']
print('\nh) *** Dataframe con registros previos al 20/12/2020:\n',antes20)

# k) Crear Dataframe pfizer con los registros en los que se haya utilizado vacuna Pfizer
country_vaccinations['vaccines'].str.split(', ')
pfizer=country_vaccinations[country_vaccinations.vaccines.str.contains('Pfizer/BioNTech')]
print('\nh) *** Dataframe con registros en los que se haya utilizado vacuna Pfizer:\n')
pfizer

# l) Almacenar los Dataframes creados en hojas individuales de Excel
#with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
#    conDiferencias.to_excel(writer,sheet_name="conDiferencias")
#    conCantidad.to_excel(writer,sheet_name="conCantidad")
#    antes20.to_excel(writer,sheet_name="antes20")
```
