c) Se agrupa por compa√±ias y se contabiliza la cantidad de vacunas

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
country_vaccinations['vaccines'].str.split()
country_vaccinations_companias=country_vaccinations['vaccines'].groupby(country_vaccinations['vaccines']).count()
print('\n*** c) Cantidad de vacunas por compa√±ias: ***\n',country_vaccinations_companias)
```

d) Se calcula el total de vacunas aplicadas a nivel mundial a partir de la columna Daily_Vaccinations

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
print('\n*** d) Total de vacunas aplicadas a nivel mundial: ',format(country_vaccinations['daily_vaccinations'].sum(), '0,.0f'))
```

e) Se calcula el promedio de vacunas aplicadas por pais

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
promedio_pais=country_vaccinations[['country','daily_vaccinations']].groupby('country').mean()
print('\n*** e) Promedio de vacunas aplicada por Pais ***',promedio_pais)
```

f) Se calcula el total de vacunas aplicadas el dia 29/01/21 a nivel mundial

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
vacunas_290121=country_vaccinations[['date','daily_vaccinations']]
vacunas_290121=vacunas_290121.query('date=="1/29/2021"')
print('\n*** f) Total de vacunas aplicadas el dia 29/01/21 a nivel mundial: ',format(vacunas_290121['daily_vaccinations'].sum(),'0,.0f'))
```

g) Se crea el Dataframe "conDiferencias" a partir del Dataframe Original

Se crea al final del Dataframe la  Columna Derivada "diferencia" con el valor obtenido de restar la columna "daily_vaccinations" de la columna"daily_vaccinations_raw"

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
conDiferencias=country_vaccinations
conDiferencias['diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
print('\n*** g) Columna Derivada con la diferencia entre campos daily_vaccinations y daily_vaccinations_raw\n', conDiferencias)
```

h) Se calcula el per√≠odo de tiempo transcurrido entre el registro con fecha mas reciente y el registro con fecha mas antigua

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
print('\n*** h) Diferencia en dias entre el registro mas reciente y el mas antiguo: ',country_vaccinations['date'].max() - country_vaccinations['date'].min())
```

i) Se crea el Dataframe conCantidad a partir del Dataframe Original

Se crea al final del Dataframe la Columna Derivada "canVac" con la cantidad de vacunas utilizadas por Farmac√©utica

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
conCantidad=country_vaccinations
conCantidad['canVac']=country_vaccinations['vaccines'].str.split(', ').str.len()
print('\n*** i) Cantidad de vacunas utilizadas por Farmac√©utica:\n',conCantidad['canVac'])
```

j) Se crea el Dataframe "antes20" con los registros previos al 20/12/2020

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
antes20=country_vaccinations[country_vaccinations['date']<'12/20/2020']
print('\n*** j) Dataframe con registros previos al 20/12/2020:\n',antes20)
```

k) Se crea el Dataframe "pfizer" con los registros en los que se haya utilizado vacuna Pfizer

```python
import pandas as pd
country_vaccinations = pd.read_csv('/content/country_vaccinations.csv', parse_dates=['date'])
country_vaccinations['vaccines'].str.split(', ')
pfizer=country_vaccinations[country_vaccinations.vaccines.str.contains('Pfizer/BioNTech')]
print('\n*** k) Dataframe con registros en los que se haya utilizado vacuna Pfizer:\n',pfizer)
```

l) Se almacena los Dataframes "conDiferencias", "conCantidad" y "antes20" en hojas individuales de Excel

```python

with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer,sheet_name="conDiferencias")
    conCantidad.to_excel(writer,sheet_name="conCantidad")
    antes20.to_excel(writer,sheet_name="antes20")
print('\n*** l) Dataframes exportados a Excel\n')
```
